

#### After having the above data, collect the data and record the number of occurrences of each neighborhood and arrange them from high to low.



```python
import pandas as pd

# Read the dataset
police_arrest = pd.read_csv("Police_Arrest_Data.csv")

# Count occurrences of each neighborhood and sort in descending order
neighborhood_counts = police_arrest['INCIDENTNEIGHBORHOOD'].value_counts().sort_values(ascending=False)

# Display top 20 neighborhoods by count
top_45_neighborhoods = neighborhood_counts.head(45)
print("Top 45 neighborhoods:")
print(top_45_neighborhoods)
```



### In order to better reflect the difference between these numbers,I created a bar chart for the top 45 neighborhoods



```python
import matplotlib.pyplot as plt

plt.figure(figsize=(15, 10))
top_45_neighborhoods.plot(kind='bar', color='red')  
plt.title('Top 45 Neighborhoods by Arrest Count')
plt.xlabel('Neighborhood') 
plt.ylabel('Number of Arrests')
plt.xticks(rotation=80)  
plt.tight_layout()
plt.show()
```
