The submetric I am researching is the overall crimerate. The dataset I used lists all the crimes and the overall crimerate that happens in each neighborhood based on the population. The submetric helps generalize the crimes that happen in each neighborhood.

### Extract and display data

```python
import pandas as pd

crime_rate = pd.read_csv("Arrests.csv")
id_ = crime_rate[['neighborhood', 'overall_crime_rate']]
print(id_)
```

### Display data in order

```python
import pandas as pd

crime_rate = pd.read_csv("Arrests.csv")
crime_rate.set_index('neighborhood', inplace=True)
hold = crime_rate['overall_crime_rate'].sort_values(ascending=False)


hold.head(25)
```

### Reflect data into a bar graph to show visual representation

```python
import pandas as pd
import matplotlib.pyplot as plt

plt.figure(figsize=(13, 8))
hold.head(40).plot(kind='bar', color='blue')  
plt.title('Top 40 Neighborhoods by overall crimerate')
plt.xlabel('Neighborhood') 
plt.ylabel('Overall crimerate')
plt.xticks(rotation=90)  
plt.tight_layout()
plt.show()
```
